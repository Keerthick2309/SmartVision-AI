```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8m.pt")
```

```python
model.train(
    data="/kaggle/input/datasets/keerthickj/smartvision-dataset/smartvision_dataset/detection/data.yaml",
    
    epochs=150,
    imgsz=768,
    batch=16,
    
    patience=30,
    
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=10,
    translate=0.1,
    scale=0.5,
    shear=2.0,
    flipud=0.0,
    fliplr=0.5,
    mosaic=1.0,
    mixup=0.2
)
```

```python
metrics = model.val()
print(metrics.box.map50)
print(metrics.box.map)
```
