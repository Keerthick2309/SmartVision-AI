```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import random
```

```python
dataset = "../smartvision_dataset/classification/train"

class_counts = {}
for class_name in os.listdir(dataset):
    class_folder = os.path.join(dataset, class_name)
    if os.path.isdir(class_folder):
        count = len(os.listdir(class_folder))
        class_counts[class_name] = count

print(class_counts)
```

```python
plt.bar(class_counts.keys(), class_counts.values())
plt.xticks(rotation=90)
plt.title("Class distribution")
plt.xlabel("Classes")
plt.ylabel("Number of Images")
plt.show()
```

```python
class_name = list(class_counts.keys())
plt.figure(figsize=(12,10))
for idx, i in enumerate(class_name):
    class_folder = os.path.join(dataset, i)
    images = os.listdir(class_folder)

    img_path = os.path.join(class_folder, random.choice(images))
    img = Image.open(img_path)

    plt.subplot(8, 4, idx + 1)
    plt.imshow(img)
    plt.title(i)
    plt.axis("off")

plt.show()
```

```python
bad_images = 0

for cl in class_name:
    class_folder = os.path.join(dataset, cl)
    for img in os.listdir(class_folder):
        try:
            Image.open(os.path.join(class_folder, img))
        except:
            bad_images += 1

print(bad_images)
```

```python
widths = []
heights = []

for class_name in os.listdir(dataset):
    class_folder = os.path.join(dataset, class_name)
    if os.path.isdir(class_folder):
        for img_file in os.listdir(class_folder)[:20]:
            img_path = os.path.join(class_folder, img_file)
            img = Image.open(img_path)

            w,h = img.size
            widths.append(w)
            heights.append(h)

print(set(widths), set(heights))
plt.hist(widths)
plt.title("Image Width Distribution")
plt.show()
```

```python
plt.hist(heights)
plt.title("Image Height Distribution")
plt.show()
```
